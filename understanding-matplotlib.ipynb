```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
x = [1,2,3]
y = [pow(a, 2) for a in x]
z = [pow(a, 3) for a in x]
plt.plot(x,y, 'o')
plt.plot(x,z)
plt.title('title of graph')
plt.xlabel('x axis')
plt.ylabel('for y & z')
# legend goes in the same order as you run the .plot commands
plt.legend(['this is y', 'this is z'])
plt.show()
```

## using Dataframes

```python
data = {'numbers': [1,2,3], 'squares':[pow(a, 2) for a in x], 'cubes': [pow(a, 3) for a in x]}
df = pd.DataFrame(data)
plt.plot(df.numbers, df.squares)
plt.plot(df.numbers, df.cubes)
plt.title('title of graph')
# plt.xlabel('x axis', fontdict={'fontsize':20, 'fontname':'Comic Sans MS'})
plt.ylabel('for y & z')
# legend goes in the same order as you run the .plot commands
plt.legend(['this is y', 'this is z'])
plt.xticks([1,2, 3])
plt.yticks([1,2,4,6,8,10])
plt.show()
```
