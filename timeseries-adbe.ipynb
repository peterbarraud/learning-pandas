```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
# convert Date col dtype to numpy (I think) datetime
parse_date = lambda x: pd.datetime.strptime(x, '%Y-%m-%d')
df = pd.read_csv('adbe.csv', parse_dates=['Date'], date_parser=parse_date, index_col='Date')
# test a date parser
# get number of rows and cols
df.shape
df.head(2)
```

```python
def get_dates_around_max(start_date, number_of_days=3):
    end_date = start_date + timedelta(days=3)
    # get the first trading day before max start
    while start_date not in df.index:
        start_date -= timedelta(days=1)
    # get 5 trading days before the first trading day before max
    five_days_before = 0
    while five_days_before < 6:
        start_date -= timedelta(days=1)
        if start_date in df.index:
            five_days_before += 1
    # get the first trading day after max end
    while end_date not in df.index:
        end_date += timedelta(days=1)
    # get 5 trading days after the first trading day after max
    five_days_after = 0
    while five_days_after < 6:
        end_date += timedelta(days=1)
        if end_date in df.index:
            five_days_after += 1
    return (start_date, end_date)
```

## MAX 2019

```python
(start_date, end_date) = get_dates_around_max(start_date=datetime(2019, 11, 2))
for row in df[start_date:end_date]['Close']:
    print(row)

```
